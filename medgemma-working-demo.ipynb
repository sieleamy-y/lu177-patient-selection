```python
# Clear GPU memory
import torch
torch.cuda.empty_cache()
import gc
gc.collect()
# Test with clinical reasoning prompt
prompt = """You are a nuclear medicine physician expert in Lu-177 PSMA radioligand therapy.

PATIENT DATA:
- Kidney absorbed dose (Ga-68 scan): 12.5 Gy
- Predicted Lu-177 kidney dose: 18.3 Gy
- eGFR: 45 mL/min/1.73m¬≤
- Prior kidney disease: Stage 3a CKD

QUESTION: Is this patient eligible for Lu-177 PSMA therapy according to EANM guidelines?

Provide structured response:
1. Eligibility (Yes/No/With modifications)
2. Key considerations
3. Dose recommendations
4. Monitoring requirements"""

inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

print("\n" + "="*70)
print("GENERATING MEDGEMMA CLINICAL RECOMMENDATION...")
print("="*70)

outputs = model.generate(
    **inputs,
    max_new_tokens=500,
    temperature=0.7,
    do_sample=True
)

response = tokenizer.decode(outputs[0], skip_special_tokens=True)

print("\n" + "="*70)
print("MEDGEMMA RESPONSE:")
print("="*70)
print(response)
```
